```python
from sklearn.linear_model import Ridge
```

```python
clf_zero = Ridge(alpha=0)
clf_zeroone = Ridge(alpha=0.01)
clf_pointone = Ridge(alpha=0.1)
clf_one = Ridge(alpha=1)
clf_ten = Ridge(alpha=10)
```

```python
acc_zero  = clf_zero.fit(X_full_train, y_train)
```

```python
acc_zeroone  = clf_zeroone.fit(X_full_train, y_train)
acc_pointone  = clf_pointone.fit(X_full_train, y_train)
acc_one  = clf_one.fit(X_full_train, y_train)
acc_ten  = clf_ten.fit(X_full_train, y_train)
```

```python
def rmse(y, y_pred):
    se = (y - y_pred) ** 2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
```

```python
X_full_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
pred_train_rr= acc_zero.predict(X_full_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_rr)))
print(r2_score(y_train, pred_train_rr))
```

```python
pred_test_rr = acc_zero.predict(X_test)
print("Score for 0 : ")
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```

```python
pred_test_rr = acc_zeroone.predict(X_test)
print("Score for 0.01 : ")
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```

```python
pred_test_rr = acc_pointone.predict(X_test)
print("Score for 0.1 : ")
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```

```python
pred_test_rr = acc_one.predict(X_test)
print("Score for 1 : ")
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```

```python
pred_test_rr = acc_ten.predict(X_test)
print("Score for 10 : ")
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```
